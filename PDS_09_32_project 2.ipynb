```python
import pandas as pd

# Load the datasets
term_test_1 = pd.read_csv(r"D:\project iict\term-test-1-result (1).csv")
term_test_2 = pd.read_csv(r"D:\project iict\term-test-2-result (1).csv")

# Display both dataframes
display(term_test_1)
display(term_test_2)

# Merge term_test_1 and term_test_2 on 'Registration Number'
merged_df = pd.merge(term_test_1, term_test_2, on="Registration Number")

# Clean up by keeping a single 'Name' column and rename the dataframe to marks
marks = merged_df.drop(columns=["Name_y"]).rename(columns={"Name_x": "Name"})

# Display the merged dataframe
print("Marks Dataframe:")
display(marks)

# Add a new column for Best Term Test Marks
marks["Best Marks"] = marks[["TT-1 Marks", "TT-2 Marks"]].max(axis=1)

# Add a new column for Average Term Test Marks
marks["Average Marks"] = marks[["TT-1 Marks", "TT-2 Marks"]].mean(axis=1)

# Display the updated dataframe
display(marks)

# Drop 'TT-1 Marks' and 'TT-2 Marks' columns
marks = marks.drop(columns=["TT-1 Marks", "TT-2 Marks"])

# Display the updated dataframe
display(marks)

Final_marks_Dataset = pd.read_csv(r"D:\project iict\final marks.csv")
Final_marks_Dataset

Final_marks_Dataset["Final Marks"] = (Final_marks_Dataset["Term Final Marks"] * 0.7) + Final_marks_Dataset["Average Marks"] + Final_marks_Dataset["Attendance Marks"]
Final_marks_Dataset

```
